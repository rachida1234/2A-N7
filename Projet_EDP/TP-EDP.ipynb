## Problème à résoudre :

\begin{cases}
-\Delta u(x, y) = f(x, y) \quad \text{sur } \Omega, \\
u(x, y) = u_d(x, y) \quad \text{sur } \partial \Omega_d, \\
\frac{\partial u(x, y)}{\partial n} = g(x, y) \quad \text{sur } \partial \Omega_n.
\end{cases}
où $\Omega = ]0,1[\times]0,1[ \subset \mathbb{R}^2$ est le domaine considéré, $\partial \Omega$ est sa frontière partitionnée en deux sous-ensembles $\partial \Omega_n$ et $\partial \Omega_d$ tels que $\partial \Omega_n \cup \partial \Omega_d = \partial \Omega$, $f \in L^2(\Omega)$, $u_d \in H^1(\Omega)$ et $g \in L^2(\partial \Omega_n)$ sont donnés.

### Ecrivons la formulation variationnelle du problème:
#### Le problème au sens des distributions s'écrira alors comme suit :
On cherche $u$ dans $H^2(\Omega)$ solution de
\begin{cases}
-\Delta u(x, y) = f(x, y) \quad \text{sur }  L^2(\Omega), \\
\gamma_0(u(x, y)) = u_d(x, y) \quad \text{sur } \partial  H^1(\Omega_d), \\
\gamma_1 (u(x, y))= g(x, y) \quad \text{sur } \partial  L^2(\Omega_n).
\end{cases}

Supposons $u$ dans $H^2(\Omega)$, et posons $v = u-u_d(x, y)$ dans $H^1_0(\Omega)$

Soit $w$ dans $L^2(\Omega)$ quelconque.

Multiplions la première équation de notre système par $w(x)$ et intégrons sur $\Omega$.

On vérifie que cette intégration est possible, compte tenu des hypothèses qui entrainent que les produits $-\Delta uw,fw$ sont intégrables sur $\Omega$. On a donc :

\begin{equation}
\begin{aligned}
    \int_\Omega \Delta u(x, y) w dx &= \int_\Omega f(x, y) w dx
\end{aligned}
\end{equation}

On a $u$ dans $H^2(\Omega)$ donc $\nabla u$ dans $H^1(\Omega)$, et pour tout $w$ dans $H^1(\Omega)$, la formule de Green nous donne :

\begin{equation}
\begin{aligned}
    \int_\Omega \nabla u \nabla w dx - \int_{\partial\Omega} \gamma_1(u) \gamma_0(w) dx &= \int_\Omega f(x, y) w dx
\end{aligned}
\end{equation}

En séparant l'intégrale : 
\begin{equation}
\begin{aligned}
    \int_\Omega \nabla u \nabla w dx - \int_{\partial\Omega_n} g \gamma_0(w) dx - \int_{\partial\Omega_d} \gamma_1(u) \gamma_0(w) dx &= \int_\Omega f(x, y) w dx
\end{aligned}
\end{equation}

Supposons maintenant que $\gamma_0(w) = 0$ sur $\partial\Omega_d$ (i.e que $w \in H^1_0(\Omega)$) 


\begin{equation}
\begin{aligned}
    \int_\Omega \nabla u \nabla w dx &= \int_\Omega f(x, y) w dx + \int_{\partial\Omega_n} g \gamma_0(w) dx
\end{aligned}
\end{equation}

On peut ainsi rechercher $u$ dans le sous espace $H^1(\Omega)$ 
En remplaçant ainsi $u = v+u_d$, on a :

\begin{equation}
\begin{aligned}
    \int_\Omega\nabla v \nabla w dx &= \int_\Omega f(x, y) w dx + \int_{\partial\Omega_n} g\gamma_0(w) dx - \int_\Omega \nabla u_d\nabla w dx
\end{aligned}
\end{equation}

#### On a ainsi le problème mis sous forme variationnelle :

\begin{aligned}
\text{Cherchons } v = u - u_d \in H_0^1(\Omega) \text{ telle que } \\
&\forall w \in H_0^1(\Omega), \int_\Omega \nabla v \cdot \nabla w dx = \int_\Omega f w dx + \int_{\partial\Omega_n} g \gamma_0(w) , dx - \int_\Omega \nabla u_d \cdot \nabla w dx \\
\text{avec } H_0^1(\Omega) := \{ w \in H^1(\Omega) \mid \gamma_0(w) = 0 \text{ sur } \partial\Omega_d \}.
\end{aligned}

### Montrons alors l'unicité de la solution :
Le problème peut s'écrire sous la forme : $a(v,w)= l(w)$ tel que $w,v \in H_0^1(\Omega) \  avec :$  
$a(v,w) = \int_\Omega \nabla v \cdot \nabla w dx \\$ 
$ l(w) = \int_\Omega f w dx + \int_{\partial\Omega_n} g \gamma_0(w) dx - \int_\Omega \nabla u_d \cdot \nabla w dx $

##### Montrons dans un premier temps que $(v,w) -> a(v,w)$ est une forme bilinéaire continue et coércive 
$*$ a est symétrique et linéaire par rapport aux deux variables donc elle est bilinéaire. 
$\\*\ a(v,w) = \int_\Omega \nabla v \cdot \nabla w dx =  <\nabla v, \nabla w >_{0,1} $
Donc en utilisant l'inégalité de Cauchy-schwartz, on aura : $ |a(v,w)| <= |v|_{0,1} |w|_{0,1} \\  $
D'où la continuité de $(v,w) -> a(v,w) \\ $
De plus, $ a(w,w) = |\nabla w|_{0,1} ^2  >= |\nabla w|_{0,1} ^2 \ $ 

On conclut ainsi par rapport à la coercivité de a.
 
##### Montrons dans un deuxième temps que $ w -> l(w)$ est une forme linéaire continue 
$*$ Par linéarité de l'intégrale et de la dérivation et sachant que $\gamma_0$ est linéaire, on peut conclure que  $ w -> l(w) $ est linéaire. 

$*$ on a : $|l(w)| = <f,w>_{L^2(\Omega)} + <g,\lambda_0(w)>_{L^2(\partial\Omega)} - < \nabla u_d,\nabla w>_{1,\Sigma} $ $ \\ $
Donc par l'inégalité triangulaire et l'inégalité de Cauchy-Schwarz on a le résultat suivant : $ \\ $
$|l(w)| < ||f||_{L^2(\Omega)}||w||_{L^2(\Omega)}+||g||_{L^2(\partial\Omega)}||\lambda_0(w)||_{L^2(\partial\Omega)} + |u_d|_{0,1}|w|_{0,1}  : (1) \\$  
D'après l'inégalité de Poincarré ($\Omega$ est un ouvert) 

Il existe $C> 0$ tel que $||w||_{L^2(\partial\Omega)} \leq C|w|_{0,1} $ $ \\ $ Et comme $\lambda_0$ est continue et linéaire alors il existe $K>0$ tel que $||\lambda_0(w)||_{L^2(\partial\Omega)} \leq K|w|_{0,1}$ et ce pour tout $v \in H_0(\Omega)$.

Ainsi, d'après l'inégalité (1), on a : $|l(w)| < L|w|_{0,1} \ $

Avec  $L = C ||f||_{L^2(\Omega)}$ + $ K ||g||_{L^2(\partial\Omega)}$ + $|u_d|_{0,1}$ $ \\ $

Et alors, $w -> l(w)$  est une forme linéaire continue. $ \\ $
A partir des conclusions faites précédemment,le théorème de Lax-Milgram peut s'appliquer et donc il existe une unique solution $v \in H_0^1(\Omega) . $ $ \\ $
Et puisque $v = u - u_d$ , il existe une unique solution u du problème initial.


### Montrons que la forme variationnelle discrète aboutit au système linéaire d'équations Ax = b :

Soit n le nombre de liberté, $ \\ $
 on peut noter w et v comme suit: 
$ \\ $
$ v_k = \sum_{i=1}^{n} v_i \eta_i $ et $ w_k = \sum_{i=1}^{n} w_i \eta_i $ qui représentent les fonctions discrètes dans la base des éléments finis avec les ($ \eta_k $) les fonctions de base des éléments finis.
$ \\ $
Ainsi, on obtient la forme variationnelle discrète suivante :
$ \\ $
On cherche $ v_k $ tq pour tout $w_k \in H_0^1(\Omega) $
 a($ v_k $,$ w_k $) = l($w_k$) $ \\ $ 
 et comme a est bilinéaire et l linéaire, on obtient l'égalité suivante :
$ \\ $
$ \sum_{i,j}v_i w_j = \sum_{i=1}^n w_i l(\eta_i) $ et ce pour tout $ w_i $
$ \\ $
ie : 
$ \sum_{i=1}^n w_i (l(\eta_i) - \sum_{j= 1}^n v_j a(\eta_j, \eta_i)) $ et ce pour tout $ w_i $
$ \\ $
donc on cherche $ v_1 $ tq pour tout $ i \in [1,n] $ : 
$ l(\eta_i) = \sum_{j=1}^n v_j a(\eta_j, \eta_i) $ 
$ \\ $
et donc en posant $A \in R^{n \times n}$ tq pour tout i,j :
$ A_{ij} = a(\eta_i,\eta_j)= \int_\Omega \nabla \eta_j \cdot \nabla \eta_i dx \\ $ 
$ \\ $
et en posant $x,b \in R^n $ tq pour tout i :
$ b_i = l(\eta_i) = \int_\Omega f \eta_i dx + \int_{\partial\Omega_n} g \gamma_0(w) dx - \int_\Omega \sum_{k= 1}^n ( U_k \nabla \eta_k) \cdot \nabla \eta_i dx $ 
et $ x = (v_j)_{j \in [1,n]} $
$ \\ $
on peut conclure que la solution qu'on cherche peut se representer en : 
$ \\ $
On cherche à trouver x tq : $Ax = b$ 
$ \\ $
La solution discrète du système linéaire est unique si la matrice A est inversible. Cette propriété est assurée si les fonctions de base $\eta_k $ sont linéairement indépendantes ce qui est bien le cas.
$ \\ $
D'où l'existence et l'unicité de la solution x du système Ax = b.

```python
def maillage(n):
#
# Une discretisation possible d'une EDP elliptique sur le domaine ]0,1[ x ]0,1[
# Le carre [0,1]x[0,1] est maille uniquement avec des triangles; 
# Les conditions limites sont de type Dirichlet uniquement   => neumann  =[];
#
# Entrees :
# n : nombre de points par cote du care => Npts points de discretisation au
# total
#
# Sorties :
# coordinates : matrice a deux colonnes. Chaque ligne contient les 
# coordonnes 2D d'un des points de la discretisation. Ces sommets seront 
# identifies a l'indice de la ligne correspondante dans la matrice
# coordinates.
# elements3 : matrice a trois colonnes. Chaque ligne contient les indices 
# des sommets d'un element triangle, dans le sens antihoraire. 
# dirichlet : vecteur colonne des indices des sommets de la frontiere de
# Dirichlet.
# neumann : matrice a deux colonnes. Chaque ligne contient les indices 
# des deux sommets d'une arete de la frontiere de Neumann.
# (neumann est vide sur cet exemple)
#
##################################################################################
    h=1/(n-1)
    npoin       = n*n ; 
    nelem       = 2*(n-1)*(n-1) ;
    coordinates = np.zeros((npoin,2)); 
    elements3   = (np.zeros((nelem,3))).astype(int) ;
    neumann     = [];
    dirichlet=(np.zeros((4*n-4,1))).astype(int)
    # Coordonnees et connectivites :
    e = -1 ; 
    p = -1 ;
    x=np.zeros((n+1,1))
    x[n,0]=1.
    for l in range (n+1):
        x[l,0]=l*h
    for j in range (n):
            for i in range(n):
                p = p + 1  
                coordinates[p,0] = x[i,0]  
                coordinates[p,1] = x[j,0] 
                if ((i != n-1) & (j != n-1)):
                    p1 = p
                    p2 = p1 + 1 
                    p3 = p1 + n 
                    p4 = p2 + n 
                    e = e + 1 
                    elements3[e,0] = p1 
                    elements3[e,1] = p2 
                    elements3[e,2] = p3 
                    e = e + 1
                    elements3[e,0] = p4 
                    elements3[e,1] = p3 
                    elements3[e,2] = p2 
    #Liste des sommets de la frontiere de Dirichlet:
    p=-1
    for j in range(n):
        p=p+1
        dirichlet[p,0] = j  
    for j in range(n*2-1,n*(n-1),n):
        p=p+1
        dirichlet[p,0] = j 
    for j in range(n*n-1,n*n-n-1,-1):
        p=p+1
        dirichlet[p,0] = j 
    for j in range(n*n-2*n,n-1,-n):
        p=p+1
        dirichlet[p,0] = j 

    return coordinates, elements3,dirichlet, neumann
```

```python
def MatriceRaideurTriangle(elements3k,coordinates):
    M = np.zeros((3,3))
    i1 = elements3k[0]
    i2 = elements3k[1]
    i3 = elements3k[2]
    Jt = np.array([[coordinates[i2,0]-coordinates[i1,0], coordinates[i3,0]-coordinates[i1,0]],
             [coordinates[i2,1]-coordinates[i1,1], coordinates[i3,1]-coordinates[i1,1]]])

    #calcul du det de Jt            
    alpha = np.linalg.det(Jt)
    for ii in range(3):
        gradeta_i = (1/alpha)*np.array([[coordinates[elements3k[(ii+1)%3],1]-coordinates[elements3k[(ii+2)%3],1]],
                                       [coordinates[elements3k[(ii+2)%3],0]-coordinates[elements3k[(ii+1)%3],0]]])
        for jj in range(3):
            gradeta_j = (1/alpha)*np.array([[coordinates[elements3k[(jj+1)%3],1]-coordinates[elements3k[(jj+2)%3],1]],
                                       [coordinates[elements3k[(jj+2)%3],0]-coordinates[elements3k[(jj+1)%3],0]]])
            #Calcul de la matrice de raideur
            M[ii,jj] = (alpha/2)*np.dot(np.transpose(gradeta_i),gradeta_j)
            
    return M,alpha              
```

```python
def Assemblage(elements3,coordinates,f):
    M = np.zeros((3,3))
    Nt = np.shape(elements3)[0]
    n = len(coordinates)
    A = np.zeros((n,n))
    b = np.zeros((n,1))
    #boucle sur le triangle    
    for k in range(len(elements3)):
        i1 = elements3[k,0]
        i2 = elements3[k,1]
        i3 = elements3[k,2]
        #Calcul du centre de masse de chaque triangle       
        xG = (coordinates[i1,0] + coordinates[i2,0] + coordinates[i3,0])/3
        yG = (coordinates[i1,1] + coordinates[i2,1] + coordinates[i3,1])/3
        M,alpha = MatriceRaideurTriangle(elements3[k,:],coordinates)
        for ii in range(3):
            for jj in range(3):
                    #Calcul de la matrice d'assemblage
                    A[elements3[k,ii],elements3[k,jj]] += M[ii,jj]

            #Assemblage du second membre
            b[elements3[k,ii]] += (alpha/6)*(f(xG,yG))
    
    
    #prise en compte des conditions aux frontières
    uh = np.zeros((n,1))
    for l in range(len(coordinates)):
        for k in range(len(coordinates)):
            if (k in dirichlet):
                uh[k] = ud(coordinates[k,0], coordinates[k,1]);
    
    #Modification de b
    b = b - A.dot(uh)
    
    #LA solution du système Auh = b à l'interieur du domaine
    vertices = [i for i in range(n)]
    #Recherche des sommets en dehors de dirichlet
    vertices_out_dirichlet = np.setdiff1d(vertices, dirichlet)
    a = A[vertices_out_dirichlet]
    Ah = a[:,vertices_out_dirichlet]
    bh = b[vertices_out_dirichlet]
    sol =  (np.linalg.inv(Ah)).dot(bh)
    uh[vertices_out_dirichlet] = sol
    
    return Ah, bh, uh
```

### Matrice de raideur pour les éléments de type quadrangle
Sachant que $$ \nabla \phi(\epsilon,\mu) = \begin{pmatrix}
\mu-1 &  \epsilon-1 \\
1-\mu &  -\epsilon \\
\mu & \epsilon \\
-\mu & 1- \epsilon
\end{pmatrix}  \ $$
Par la formule du changement de variable, on peut conclure :
$M_{ij} = \int_{0,1}\int_{0,1} \nabla \phi_i^T(J_\phi^TJ_\phi)^{-1} \nabla \phi_j^T|J_\phi|d\epsilon d\mu \\ $ 
avec $J_\phi$ la matrice jacobienne de $\phi$
et comme $J_\phi^TJ_\phi$ est symetrique alors son inverse l'est aussi soit donc la matrice suivante : 
$$ (J_\phi^TJ_\phi)^{-1} = \begin{pmatrix}
a &  b \\
b & c 
\end{pmatrix}  \ $$
Par calcul des intégrales:
$$
M_{0,0} = \int_0^1 \int_0^1 (y-1)(a(y-1)+b(x-1)) +(x-1)(b(y-1)+c(x-1)) dx dy 
\\
= \frac{1}{6}(2a+3b+2c)
$$
 on obtient :
$$ M =  \frac{|J_\phi|}{6} \begin{matrix}
 2a+3b+2c & -2a+c & -a-3b-c & a-2c \\
-2a+c & 2a-3b+2c & a-2c & -a+3b-c \\
-a-3b-c & a-2c & 2a+3b+2c & -2a+c \\
a-2c & -a+3b-c & -2a+c & 2a-3b+2c
\end{matrix}  $$ 


```python
def Jphik(coordinates,elements4k):
    i1 = elements4k[0]
    i2 = elements4k[1]
    i3 = elements4k[2]
    i4 = elements4k[3]
    Jphik = np.array([[coordinates[i2,0]-coordinates[i1,0], coordinates[i4,0]-coordinates[i1,0]],
             [coordinates[i2,1]-coordinates[i1,1], coordinates[i4,1]-coordinates[i1,1]]])
        
        
    return Jphik
```

```python
def MatriceRaideurQuadrangle(elements4k,coordinates):
    M = np.zeros((4,4))
    n = len(coordinates)
    J = Jphik(coordinates,elements4k)
    Jnorm = np.linalg.norm(J)
    alpha = np.linalg.det(J)
    Jinv = np.linalg.inv(np.transpose(J)*J)
    a = Jinv[0,0]
    b = Jinv[0,1]
    c = Jinv[1,1]

    #Calcul de la matrice de raideur
    M[0,0] = (2*a+3*b+2*c)
    M[1,1] = (2*a-3*b+2*c)
    M[2,2] = (2*a+3*b+2*c)
    M[3,3] = (2*a-3*b+2*c)
    M[0,1] = (-2*a+c)
    M[1,2] = (a-2*c)
    M[2,3] = (-2*a+c)
    M[0,2] = (-a-3*b-c)
    M[1,3] = (-a+3*b-c)
    M[0,3] = (a-2*c)
    #Calcul de la matrice d'assemblage
    M = (Jnorm/6)*(np.transpose(M) + M - np.diag(M)*np.eye(4,4))
    
    return M, alpha 
    
```

```python
def AssemblageMixte(coordinates,elements4,elements3,f,g,dirichlet,neumann):
    n = len(coordinates)
    A = np.zeros((n,n))
    b = np.zeros((n,1)) 
    
    #Assemblage quadrangulaire
    for k in range(len(elements4)):
        i1 = elements4[k][0]
        i2 = elements4[k][1]
        i3 = elements4[k][2]
        i4 = elements4[k][3]
        #Calcul du centre de masse de chaque triangle       
        xG = (coordinates[i1,0] + coordinates[i2,0] + coordinates[i3,0]+ coordinates[i4,0])/4
        yG = (coordinates[i1,1] + coordinates[i2,1] + coordinates[i3,1]+ coordinates[i4,1])/4
        MQuadrangle,alphaQ = MatriceRaideurQuadrangle(elements4[k,:],coordinates)
        for ii in range(4):
            for jj in range(4):
                A[elements4[k,ii],elements4[k,jj]] += MQuadrangle[ii,jj]
                
            #Assemblage du second membre
            b[elements4[k,ii]] += (alphaQ/4)*(g(xG,yG))
        
    #Assemblage triangulaire   
    for k in range(len(elements3)):
        i1 = elements3[k][0]
        i2 = elements3[k][1]
        i3 = elements3[k][2]
        Mtriangle,alphaT = MatriceRaideurTriangle(elements3[k,:],coordinates)
        #Calcul du centre de masse de chaque triangle       
        xG = (coordinates[i1,0] + coordinates[i2,0] + coordinates[i3,0])/3
        yG = (coordinates[i1,1] + coordinates[i2,1] + coordinates[i3,1])/3
        for ii in range(3):
            for jj in range(3):
                 A[elements3[k,ii],elements3[k,jj]] += Mtriangle[ii,jj]
            #Assemblage du second membre
            b[elements3[k,ii]] += (alphaT/6)*(f(xG,yG))    
    

    
    #prise en compte des conditions aux frontières
    uh = np.zeros((n,1))
    for l in range(len(coordinates)):
        for k in range(len(coordinates)):
            if (k in dirichlet):
                uh[k] = ud(coordinates[k,0], coordinates[k,1]);
    #Modification de b
    b = b - A.dot(uh)
                
    #prise en compte des conditions de neumann
    for l in range(len(neumann)):
        (i,j) = neumann[l,:] ;
        [sx0,sy0] = coordinates[i];
        [sx1,sy1] = coordinates[j];
        b[i] += (1/2)*math.sqrt((sx0-sx1)**2 +(sy0-sy1)**2)*g((sx0+sx1)/2,(sy0+sy1)/2)
        b[j] += (1/2)*math.sqrt((sx0-sx1)**2 +(sy0-sy1)**2)*g((sx0+sx1)/2,(sy0+sy1)/2)

    #La résolution du système Auh = b à l'interieur du domaine
    vertices = [i for i in range(n)]
    #Recherche des sommets en dehors de dirichlet
    vertices_out_dirichlet = np.setdiff1d(vertices, dirichlet)
    a = A[vertices_out_dirichlet]
    Ah = a[:,vertices_out_dirichlet]
    bh = b[vertices_out_dirichlet]
    sol =  (np.linalg.inv(Ah)).dot(bh)
    uh[vertices_out_dirichlet] = sol
    
    return Ah,bh,uh
```

## Problème à résoudre :

\begin{cases}
-\Delta u(x, y) + c_0 u(x,y) = f(x, y) \quad \text{sur } \Omega, \\
u(x, y) = 0 \quad \text{sur } \partial \Omega, \\
\end{cases}
où $\Omega = ]0,1[\times]0,1[ \subset \mathbb{R}^2$ est le domaine considéré, et c_0 est une constante

### Ecrivons la formulation variationnelle du problème:
#### Le problème au sens des distributions s'écrira alors comme suit :
On cherche $u$ dans $H^2(\Omega) \cap H_0^1(\Omega) $ solution de
\begin{cases}
-\Delta u(x, y) +  c_0 u(x,y) = f(x, y) \quad \text{sur }  L^2(\Omega), \\
\gamma_0(u(x, y)) = 0 \quad \text{sur } \partial \Omega, \\
\end{cases}


Soit $v$ dans $L^2(\Omega)$ quelconque.

Multiplions la première équation de notre système par $v(x)$ et intégrons sur $\Omega$.

On vérifie que cette intégration est possible, compte tenu des hypothèses qui entrainent que les produits $-\Delta uv,fv$ sont intégrables sur $\Omega$. On a donc :

\begin{equation}
\begin{aligned}
    \int_\Omega \Delta u(x, y) v dx + c_0 \int_\Omega u(x, y) v dx &= \int_\Omega f(x, y) v dx
\end{aligned}
\end{equation}

On a $u$ dans $H^2(\Omega)$ donc $\nabla u$ dans $H^1(\Omega)$, et pour tout $v$ dans $H^1(\Omega)$, la formule de Green nous donne :

\begin{equation}
\begin{aligned}
    \int_\Omega \nabla u \nabla v dx - \int_{\partial\Omega} \gamma_1(u) \gamma_0(v) dx + c_0 \int_\Omega u(x, y) v dx &= \int_\Omega f(x, y) v dx
\end{aligned}
\end{equation}

Supposons maintenant que $\gamma_0(v) = 0$ sur $\partial\Omega$ (i.e que $v \in H^1_0(\Omega)$) 


\begin{equation}
\begin{aligned}
    \int_\Omega \nabla u \nabla v dx + c_0 \int_\Omega u(x, y) v dx &= \int_\Omega f(x, y) v dx
\end{aligned}
\end{equation}
On peut donc chercher u dans le sous espace $ H^1_0(\Omega)$

#### On a ainsi le problème mis sous forme variationnelle :

\begin{aligned}
\text{Cherchons } u \in H_0^1(\Omega) \text{ tel que } \\
&\forall v \in H_0^1(\Omega), \int_\Omega \nabla u \cdot \nabla v dx + c_0 \int_\Omega u(x, y) v dx = \int_\Omega f v dx \\
\text{avec } H_0^1(\Omega) := \{ v \in H^1(\Omega) \mid \gamma_0(v) = 0 \text{ sur } \partial\Omega \}.
\end{aligned}



### Montrons alors l'unicité de la solution : 
Le problème peut s'écrire sous la forme : $a(u,v)= l(v)$ tel que $u,v \in H_0^1(\Omega) \  avec :$  
$a(v,w) = \int_\Omega \nabla v \cdot \nabla w dx + c_0 \int_\Omega uv dx \\$ 
$ l(w) = \int_\Omega f v dx $

##### Montrons dans un premier temps que $(u,v) -> a(u,v)$ est une forme bilinéaire continue et coércive 
$*$ a est symétrique et linéaire par rapport aux deux variables donc elle est bilinéaire. 
$\\*\ a(u,v) = \int_\Omega \nabla u \cdot \nabla v dx + c_0 \int_\Omega uv dx =  <\nabla v, \nabla w >_{0,1} + c_0 <u,v>_{L^2(\Omega)} \\ $
Donc en utilisant l'inégalité de Cauchy-schwartz ($(H_0^1(\Omega),<.,.>_{0,1})$ et $(L^2(\Omega),<.,.>_{L^2(\Omega)}$ sont des Hilberts)),et comme $c_0$>0, on aura : $ |a(u,v)| \leq |u|_{0,1} |v|_{0,1} + c_0 ||u||_{L^2(\Omega)} ||v||_{L^2(\Omega)} \\  $
D'après l'inégalité de Poincarré ($\Omega$ est un ouvert) 

Il existe $C> 0$ tel que $||v||_{L^2(\Omega)} \leq C |v|_{0,1} $ $ \\ $
et donc $|a(u,v) \leq (1+c_0C)|u|_{0,1} |v|_{0,1} $
D'où la continuité de $(u,v) -> a(u,v) \\ $
De plus, $ |a(v,v)| = |\int_\Omega \nabla v^2 dx +c_0\int\Omega v^2| \ $
ie : $|a(v,v)| =  |v|^2_{0,1} + c_0||v||^2_{L^2(\Omega)}\ >= |v|^2_{0,1} +c_0|v|^2_{0,1} \ $ ainsi :
$|a(v,v)| >= (1+c_0)|v|^2_{0,1}$

On conclut ainsi par rapport à la coercivité de a.
 
##### Montrons dans un deuxième temps que $ v -> l(v)$ est une forme linéaire continue 
$*$ Par linéarité de l'intégrale et de la dérivation , on peut conclure que  $ v -> l(v) $ est linéaire. 

$*$ on a : $|l(v)| = <f,v>_{L^2(\Omega)} $ $ \\ $
Donc par l'inégalité de Cauchy-Schwarz on a le résultat suivant : $ \\ $
$|l(v)| < ||f||_{L^2(\Omega)}||v||_{L^2(\Omega)} : (1) \\ $  
et comme $\Omega$ est un ouvert on peut utiliser l'ingalité de Poincaré, d'où l'existence de C>0 tq : 
$||v||_{L^2(\Omega)}  \leq C|v|_{0,1} \\ $

Ainsi, d'après l'inégalité (1), on a :
 $|l(v)| < L|v|_{0,1} \ $

$Avec  \ L = C ||f||_{L^2(\Omega)} \\ $

Et alors, $v -> l(v)$  est une forme linéaire continue. $ \\ $
A partir des conclusions faites précédemment,le théorème de Lax-Milgram peut s'appliquer et donc il existe une unique solution $u \in H_0^1(\Omega) . $ $ \\ $








### Montrons que la forme variationnelle discrète aboutit au système linéaire d'équations Ax = b :

Soit n le nombre de liberté, $ \\ $
 on peut noter w et v comme suit: 
$ \\ $
$v_k = \sum_{i=1}^{n} v_i \eta_i $ et $w_k = \sum_{i=1}^{n} w_i \eta_i $ qui représentent les fonctions discrètes dans la base des éléments finis avec les ($\eta_k$) les fonctions de base des éléments finis.
$ \\ $
Ainsi, on obtient la forme variationnelle discrète suivante :
$ \\ $
On cherche $v_k$ tq pour tout $w_k \in H_0^1(\Omega) $
 a($v_k$,$w_k$) = l($w_k$) $ \\ $ 
 et comme a est bilinéaire et l linéaire, on obtient l'égalité suivante :
$ \\ $
 $\sum_{i,j}v_i w_j = \sum_{i=1}^n w_i l(\eta_i)$ et ce pour tout $ w_i $
$ \\ $
ie : 
$\sum_{i=1}^n w_i (l(\eta_i) - \sum_{j= 1}^n v_j a(\eta_j, \eta_i))$ et ce pour tout $ w_i $
$ \\ $
donc on cherche $v_1$ tq pour tout $i \in [1,n] $ : 
$l(\eta_i) = \sum_{j=1}^n v_j a(\eta_j, \eta_i) $ 
$ \\ $
et donc en posant $A \in R^{n \times n}$ tq pour tout i,j :
$ A_{ij} = a(\eta_i,\eta_j)= \int_\Omega \nabla \eta_j \cdot \nabla \eta_i dx  +  c_0 \int_\Omega  \eta_i  \eta_j  dx \\ $ 
$ \\ $
et en posant $x,b \in R^n $ tq pour tout i :
$ b_i = l(\eta_i) = \int_\Omega f \eta_i dx $
et $ x = (v_j)_{j \in [1,n]} $
$ \\ $
on peut conclure que la solution qu'on cherche peut se representer en : 
$ \\ $
On cherche à trouver x tq : $Ax = b$ 
$ \\ $
La solution discrète du système linéaire est unique si la matrice A est inversible. Cette propriété est assurée si les fonctions de base $\eta_k $ sont linéairement indépendantes ce qui est bien le cas.
$ \\ $
D'où l'existence et l'unicité de la solution x du système Ax = b.


###  La stratégie de l’élément de référence pour le calcul de l'integrale :

### Résolution numérique de la formulation variationnelle discrète :

```python
def solve_discrete_variational_formulation(elements3, coordinates, dirichlet):
    # initialize matrix A and vector b
    n = len(coordinates)
    A = np.zeros((n, n))
    b = np.zeros((n, 1))
    
    # define the function f(x,y)
    def f(x,y):
        return 1

    # assemble the global stiffness matrix
    for i in range(len(elements3)):
        # get the vertices of the current element
        vertices = [coordinates[elements3[i][0]], coordinates[elements3[i][1]], coordinates[elements3[i][2]]]
        # calculate the element stiffness matrix and coefficient alpha
        K, alpha = MatriceRaideurTriangle(elements3[i,:],coordinates)
        # assemble the global stiffness matrix
        # calculate the centroid of the element
        centroid = (1/3) * (vertices[0] + vertices[1] + vertices[2])
        # calculate the area of the element
        area = 0.5 * abs(vertices[0][0]*vertices[1][1] + vertices[1][0]*vertices[2][1] + vertices[2][0]*vertices[0][1] - vertices[1][0]*vertices[0][1] - vertices[2][0]*vertices[1][1] - vertices[0][0]*vertices[2][1])
        c0 = 5
        for j in range(3):
            for k in range(3):
                if(j == k):
                    A[elements3[i][j]][elements3[i][k]] += c0 * area /12 
                else:
                    A[elements3[i][j]][elements3[i][k]] += c0 * area /24
            # approximate the integral of the load term over the element
            b[elements3[i][j]] += (alpha/6) * f(centroid[0],centroid[1])

   
    
    ld, cd = np.shape(dirichlet)
    u_h = np.zeros((n,1))
    # apply Dirichlet boundary conditions
    def ud(x,y):
        return 0
    
    for i in range(ld):
        id_vertex = dirichlet[i,0]
        x,y = coordinates[id_vertex]
        u_h[id_vertex] = ud(x,y)
    
    # modify vector b due to Dirichlet boundary conditions
    b = b - A.dot(u_h)
    
    # solve for u on the interior points
    interior_vertices = np.setdiff1d(np.arange(n), dirichlet)
    A_interior = A[interior_vertices][:,interior_vertices]
    b_interior = b[interior_vertices]
    u_interior = np.linalg.solve(A_interior, b_interior)
    u_h[interior_vertices] = u_interior 
    
    return  A_interior, b_interior,u_h
```
